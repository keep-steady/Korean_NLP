pyLDAvis는 LDA모델의 학습 결과를 시각적으로 확인하는데 유용한 툴킷임. 

```python
import pickle
from scipy.io import mmread
import gensim
import pyLDAvis

mm_fname = '/home/paulkim/workspace/python/Korean_NLP/data/corpus_10days/models/2016-10-24_article_all_normed_corpus.mtx'
dictionary_fname = '/home/paulkim/workspace/python/Korean_NLP/data/corpus_10days/models/2016-10-24_article_all_normed_corpus.dictionary'
ldamodel_fname = '/home/paulkim/workspace/python/Korean_NLP/data/corpus_10days/models/2016-10-24_article_all_normed_corpus_lda.pkl'
pyldavis_fname = '/home/paulkim/workspace/python/Korean_NLP/data/corpus_10days/models/2016-10-24_article_all_normed_corpus_pyldavis.pkl'

TRAIN_PYLDAVIS = True
```

```python
if TRAIN_PYLDAVIS:
    dictionary = gensim.corpora.Dictionary.load_from_text(dictionary_fname)
    corpus = gensim.matutils.Sparse2Corpus(mmread(mm_fname).transpose())
    with open(ldamodel_fname, 'rb') as f:
        lda = pickle.load(f)
        
else:
    with open(ldamodel_fname, 'rb') as f:
        lda = pickle.load(f)
```

```python
lda.show_topics(num_words=5)
```

pyLDAvis.gensim.prepare()를 이용하면 시각화 할 준비를 함

```python
if TRAIN_PYLDAVIS:
    import pyLDAvis.gensim
    pyldavis = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

pyLDAvis.display()를 하면 interactive display가 됨

relevance(term w | topic t) = $\lambda$ p(w | t) + ( 1 - $\lambda$) p(w | t)p(w)

```python
pyLDAvis.display(pyldavis)
```

pyldavis 역시 picking이 가능함

```python
if TRAIN_PYLDAVIS:
    with open(pyldavis_fname, 'wb') as f:
        pickle.dump(pyldavis, f)
```

```python
with open(pyldavis_fname, 'rb') as f:
    loaded_pyldavis = pickle.load(f)
```

```python
pyLDAvis.display(loaded_pyldavis)
```
